```python
def gpa_info(yr):
    gpas_yr = gpas.loc[gpas['Fall Term'] == yr]
    gpas_yr_adm = gpas_yr.loc[gpas_yr['Measure Names'].isin(['Adm GPA'])]
    
    gpas_yr_adm_campus = []

    for c in campus_names:
        gpas_yr_adm_campus.append(gpas_yr_adm.loc[gpas_yr_adm['Campus'].isin([c])])
    
    gpas_yr_adm_campus_m = []

    for s in gpas_yr_adm_campus:
        gpas_yr_adm_campus_m.append(np.mean(s['Measure Values']))
    
    return gpas_yr_adm_campus_m
```

```python
yr = np.arange(1994, 2018)
gpa_yr_campus = [[], [], [], [], [], [], [], [], [], []]
campus_names = ['Berkeley', 'Los Angeles', 'San Diego', 'Santa Barbara', 'Irvine', 
                'Universitywide', 'Davis', 'Santa Cruz', 'Riverside', 'Merced']
colors = ['red', 'green', 'yellow', 'blue', 'black', 'gray', 'purple', 'brown', 'cyan', 'orange']

for y in yr:
    for i in range(0, len(gpa_yr_campus)):
        gpa_yr_campus[i].append(gpa_info(y)[i])

plt.ylabel('GPA')
plt.xlabel('Year')
plt.title('GPA Inflation Over the Years')
for k in range(0, len(gpa_yr_campus)):
    plt.plot(yr, gpa_yr_campus[k], '-o', color=colors[k])
```
